# Label Encoding

```python
# label encode categorical columns
from sklearn.preprocessing import LabelEncoder
lblEncoder_state = LabelEncoder()
lblEncoder_state.fit(data['STATE'])
data['STATE'] = lblEncoder_state.transform(data['STATE'])

lblEncoder_cons = LabelEncoder()
lblEncoder_cons.fit(data['CONSTITUENCY'])
data['CONSTITUENCY'] = lblEncoder_cons.transform(data['CONSTITUENCY'])

lblEncoder_name = LabelEncoder()
lblEncoder_name.fit(data['NAME'])
data['NAME'] = lblEncoder_name.transform(data['NAME'])

lblEncoder_party = LabelEncoder()
lblEncoder_party.fit(data['PARTY'])
data['PARTY'] = lblEncoder_party.transform(data['PARTY'])

lblEncoder_gender = LabelEncoder()
lblEncoder_gender.fit(data['GENDER'])
data['GENDER'] = lblEncoder_gender.transform(data['GENDER'])

lblEncoder_category = LabelEncoder()
lblEncoder_category.fit(data['CATEGORY'])
data['CATEGORY'] = lblEncoder_category.transform(data['CATEGORY'])

lblEncoder_edu = LabelEncoder()
lblEncoder_edu.fit(data['EDUCATION'])
data['EDUCATION'] = lblEncoder_edu.transform(data['EDUCATION'])
data
```

```python
# clean ASSETS and LIABILITIES column values
data['ASSETS'] = data['ASSETS'].map(lambda x: x.lstrip('Rs ').split('\n')[0].replace(',', ''))
data['LIABILITIES'] = data['LIABILITIES'].map(lambda x: x.lstrip('Rs ').split('\n')[0].replace(',', ''))
data.head()
```

```python
# replace Nil values with 0
data['ASSETS'] = data['ASSETS'].replace(['Nil', '`', 'Not Available'], '0')
data['LIABILITIES'] = data['LIABILITIES'].replace(['NIL', '`', 'Not Available'], '0')
data['CRIMINAL_CASES'] = data['CRIMINAL_CASES'].replace(['Not Available'], '0')
# convert ASSETS and LIABILITIES columns values into numeric
data['ASSETS'] = data['ASSETS'].astype(str).astype(float)
data['LIABILITIES'] = data['LIABILITIES'].astype(str).astype(float)
```

# Predictive Models

```python
data['WINNER'].value_counts()
```

```python
data['WINNER'].value_counts(normalize=True)
```

```python
#spliting the dataset 
X=data.drop(['WINNER'],axis=1)
y=data['WINNER']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.shape
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(X_train,y_train)
y_pred=logit_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
print("Accuracy is:",accuracy_score(y_test,y_pred))
print("Precision is:",precision_score(y_test,y_pred))
print("Recall is:",recall_score(y_test,y_pred))
print("f1_score is:",f1_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

There are 22 misclassifications

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
acc_values=[]
neighbors=np.arange(3,15)
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski')
    classifier.fit(X_train,y_train)
    y_pred=classifier.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    acc_values.append(acc)
```

```python
acc_values
```

```python
plt.plot(neighbors,acc_values,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
```

```python
classifier=KNeighborsClassifier(n_neighbors=3,metric='minkowski')
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
```

```python
print("Accuracy is:",accuracy_score(y_test,y_pred))
print("Precision is:",precision_score(y_test,y_pred))
print("Recall is:",recall_score(y_test,y_pred))
print("f1_score is:",f1_score(y_test,y_pred))
```
